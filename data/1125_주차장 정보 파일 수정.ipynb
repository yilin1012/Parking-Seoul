## 11월 25일 주차장 안내정보 데이터 수정

### 문제 발생 : PK가 없다.
### 문제 발생 원인 : 주차장코드는 PK 코드가 될 수 있다. 
###                         주차장 위도, 경도 변수가 주차장 자리 하나마다 할당되어 저장된 데이터들이 있다.
### 문제 해결 방법 : 주차장 위도, 경도 변수를 삭제한다.
### 문제 해결 방법2 : 데이터프레임에서 중복되는 행을 삭제한다.
### 문제 해결 완료! : 주차장코드를 PK로 삼는다.

```python
# 데이터 불러오기
import pandas as pd
data=pd.read_csv('/Users/yerim/Desktop/parkingseoul/data/서울시 공영주차장 안내정보_편집본(1).csv', encoding='euc-kr')
data.head()
```

```python
# 데이터 정보 확인
data.info()
```

```python
# 주차장 위치 좌표 위도와 경도 삭제
del data['주차장 위치 좌표 위도']
del data['주차장 위치 좌표 경도']
del data['주말 운영 시작시각(HHMM)']
del data['주말 운영 종료시각(HHMM)']
del data['공휴일 운영 시작시각(HHMM)']
del data['공휴일 운영 종료시각(HHMM)']
del data['토요일 유,무료 구분']
del data['토요일 유,무료 구분명']
del data['공휴일 유,무료 구분']
del data['공휴일 유,무료 구분명']
```

```python
# 위도 경도 변수 삭제 후 데이터 확인
data.info()
```

```python
# 데이터프레임에서 중복되는 행 제거
data = data.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=True)
```

```python
# 데이터프레임에서 중복되는 행 제거후 데이터 확인
data.info()
```

```python
# 중복되는 행 제거, 위도, 경도 변수 삭제 후 주차장 코드 확인
print(data['주차장코드'].value_counts())
print(data['주차장코드'].value_counts().sum())
```

```python
# 이제 주차장 코드는 PK가 되었다.
code1 = data[data['주차장코드']== 1535968]
code1.head()
```

```python
data.to_csv('주차장 안내 정보 파일_최종.csv',  encoding='euc-kr', index=False)
```
