### 서울시 공영주차장 안내정보 요약 파일 토대로 서울시 공영주차장 안내정보 csv 수정

#### 주요 수정 내용
##### (1) 주소 속성에서 00구를 기준으로 문자열 split 해서 "구" 정보 따로 만들기 ✔️
##### (2) 주소 속성에서 00동을 기준으로 문자열 split 해서 "동" 정보 따로 만들기 ✔️
##### (3) 운영구분에서 이륜차 전용 주차장, 버스전용 주차장 행은 삭제하기 ✔️
##### (4) 월 정기권 금액 열 삭제하기 ✔️
##### (5) 노상 주차장 관리 그룹 번호 열 삭제하기 ✔️
##### (6) 버스 기본 주차 요금 , 버스 기분 주차 시간(분 단위), 버스 추가 단위 요금, 버스 추가 단위 시간(분 단위) 열 삭제하기 
##### '야간무료개방여부', '야간무료개방여부명',주말 운영 시작시각(HHMM)','주말 운영 종료시각(HHMM)',
##### '공휴일 운영 시작시각(HHMM)','공휴일 운영 종료시각(HHMM)','토요일 유,무료 구분','토요일 유,무료 구분명',
##### '공휴일 유,무료 구분','공휴일 유,무료 구분명', 
##### 열 삭제
##### (7) 총 주자면 -> 종 주차면으로 오타 수정 ✔️

```python
# 데이터 불러오기
import pandas as pd
data=pd.read_csv('/Users/yerim/Desktop/parkingseoul/서울시 공영주차장 안내 정보.csv', encoding='euc-kr')
```

```python
# 필요하지 않은 열 삭제하기 ; 노상 주차장 관리 그룹 번호, 버스 기본 주차 요금, 버스 기본 주차 시간(분 단위), 버스 추가 단위 요금, 버스 추가 단위 사간(분 단위) 
new_data = data[['주차장코드','주차장명','주소','주차장 종류','주차장 종류명',
                 '운영구분','운영구분명','전화번호','총 주자면','유무료구분',
                 '유무료구분명','야간무료개방여부', '야간무료개방여부명',
                 '평일 운영 시작시각(HHMM)','평일 운영 종료시각(HHMM)',
                 '최종데이터 동기화 시간','토요일 유,무료 구분','토요일 유,무료 구분명',
                 '기본 주차 요금','기본 주차 시간(분 단위)',
                 '추가 단위 요금','추가 단위 시간(분 단위)','일 최대 요금','주차장 위치 좌표 위도','주차장 위치 좌표 경도']]

new_data.info()
```

```python
# 오타 수정! 총 주자면 열 -> 총 주차면
new_data = new_data.rename(columns={'총 주자면':'총 주차면'})

new_data.head()
```

```python
# 구 열 추가
new_data['구']=new_data['주소']

# 구 열 만들기
for i in range(8000) :
    try : 
        new_data['구'][i]=new_data['구'][i].split()[0]
    except :
        new_data['구'][i]=None
```

```python
# 구 열 확인
print(new_data['구'].value_counts())
print(new_data['구'].value_counts().sum())
```

```python
# 행정동 열 추가
new_data['행정동']=new_data['주소']

# 구 열 만들기
for i in range(8000) :
    try : 
        new_data['행정동'][i]=new_data['행정동'][i].split()[1]
    except :
        new_data['행정동'][i]=None
```

```python
# 행정동 열 확인
print(new_data['행정동'].value_counts())
print(new_data['행정동'].value_counts().sum())

# 성수동2가 처럼 행정동과 상세주소지가 붙어있는 경우 있지만, '동'을 기준으로 문자열을 자르는 것이 번거로우므로 ex) 동선동의 경우 '동'으로 문자열 찾을 수 없음.. 
# 일단 내버려두기~!
```

```python
# 운영 구분명이 '이륜차 전용 주차장', '버스전용 주차장' 인 행 

delete = new_data[(new_data['운영구분명']=='이륜차 전용 주차장')|(new_data['운영구분명']=='버스전용 주차장')]
delete

# 총 353개
```
